```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('seaborn')
```

```python
train_df = pd.read_csv('../input/train.csv')
test_df = pd.read_csv('../input/test.csv')
```

```python
train_df.to_parquet('./tmp_data/train_df.parquet')
test_df.to_parquet('./tmp_data/test_df.parquet')
```

```python
features = [c for c in train_df.columns if c not in ['ID_code', 'target']]
```

```python
target = train_df['target']
train = train_df.drop(["ID_code", "target"], axis=1)
sns.set_style('whitegrid')
sns.countplot(target)
```

```python
plt.figure(figsize=(16,6))
plt.title("Distribution of mean values per row in the train and test set")
sns.distplot(train_df[features].mean(axis=1),color="black", kde=True,bins=120, label='train')
sns.distplot(test_df[features].mean(axis=1),color="red", kde=True,bins=120, label='test')
plt.legend()
plt.show()
```

```python
corrmat = train_df[features + ['target']].corr(method='spearman')
f, ax = plt.subplots(figsize=(50, 40))
sns.heatmap(corrmat, ax=ax, linewidths=0.1)
```
